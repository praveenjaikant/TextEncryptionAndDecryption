Supposing we receive an encrypted message. The function decrypt that takes in a message, and produces all possible translations of the message using a Caesar cipher. The performance of this function has been  demonstrated on the following encrypted quotes from Greek philosophers:

1. Aristotle: Swcnkva ku pqv cp cev, kv ku c jcdkv.

2. Demosthenes: Dxlww zaazcefytetpd lcp zqepy esp mprtyytyr zq rcple pyepcactdpd.

3. Hypatia: Fczy cm uh ohzifxgyhn, uhx nby zolnbyl qy nlupyf nby gily nlonb qy wuh wigjlybyhx. Ni ohxylmnuhx nby nbcham nbun uly un iol xiil cm nby vymn jlyjuluncih zil ohxylmnuhxcha nbimy nbun fcy vysihx.

The shift values are in the range of  [-12, 13]

```python
def decrypt (string):
    decrypt_dict = {}
    print('\nDecrypting the message:')
    for shift in range(-12,14):
        shifted_string = ''
        for char in string:
            char_ascii = ord(char) # using the ord() function to get the ascii value
            char_ascii_shifted = char_ascii + shift # shifting the value to the right
            if (char_ascii >= 65 and char_ascii <= 90): # checking if imput character is upper case
                if (char_ascii_shifted > 90): # if, after shifting, the ascii value has exceeded that of 'Z'
                    char_ascii_shifted_back = 65 + ((char_ascii_shifted - 1) % 90) # then restart from 'A' + shift
                    char_right_shifted = chr(char_ascii_shifted_back) # convert back from ascii value to character
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                    
                elif ((char_ascii_shifted >= 65 and char_ascii_shifted <= 90)): # if ascii value is within limits of uppercase
                    char_right_shifted = chr(char_ascii_shifted) #convert ascii value to character value
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                    
                elif (char_ascii_shifted < 65): # if, after shifting, the value has gone below 'A'
                    char_ascii_shifted_back = 90 - ((65 - char_ascii_shifted) - 1) # then shift to 'Z' and backwards
                    char_right_shifted = chr(char_ascii_shifted_back) # convert back from ascii value to character
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                    
            elif (char_ascii >= 97 and char_ascii <= 122): # else if the character is in lowercase
                char_ascii_shifted = char_ascii + shift # shifting the value to the right
                if (char_ascii_shifted > 122): # if, after shifting, the ascii value has exceeded that of 'z'
                    char_ascii_shifted_back = 97 + ((char_ascii_shifted - 1) % 122) # then restart from 'a' + shift
                    char_right_shifted = chr(char_ascii_shifted_back) # convert back from ascii value to character
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                
                elif ((char_ascii_shifted >= 97 and char_ascii_shifted <= 122)): # if ascii value is within limits of lowercase
                    char_right_shifted = chr(char_ascii_shifted) #convert ascii value to character value
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                    
                elif (char_ascii_shifted < 97): # if, after shifting, the value has gone below 'a'
                    char_ascii_shifted_back = 122 - ((97 - char_ascii_shifted) - 1) # then, go to 'z' and start going backwards
                    char_right_shifted = chr(char_ascii_shifted_back) # convert back from ascii value to character
                    shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
                
            else:
                char_right_shifted = char # if non-alphabet, keep character as it is
                shifted_string = shifted_string + char_right_shifted # appending shifted character to final string
        
        print('shift = ',shift)
        print('decrypted message = ', shifted_string,'/n')
        decrypt_dict[shift] = shifted_string
        # decrypted string is stored as value, shift value is stored as key in final dictionary
    return decrypt_dict

Aristotle = decrypt('Swcnkva ku pqv cp cev, kv ku c jcdkv')
print ('\ncorrect message 1 = ', Aristotle[-2])
Demosthenes = decrypt('Dxlww zaazcefytetpd lcp zqepy esp mprtyytyr zq rcple pyepcactdpd')
print ('\ncorrect message 2 = ', Demosthenes[-11])
Hypatia = decrypt('Fczy cm uh ohzifxgyhn, uhx nby zolnbyl qy nlupyf nby gily nlonb qy wuh wigjlybyhx. Ni ohxylmnuhx nby nbcham nbun uly un iol xiil cm nby vymn jlyjuluncih zil ohxylmnuhxcha nbimy nbun fcy vysihx')
print ('\ncorrect message 3 = ', Hypatia[6])
```
